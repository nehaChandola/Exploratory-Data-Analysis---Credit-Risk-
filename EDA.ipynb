### Univariate Analysis

#### Categorical Variable

```python
def plot_categorical_uni_bi(feature):
    fig, ax = plt.subplots(1,2,figsize=(15,5))
    sns.countplot(x=feature,data=df,palette="Oranges_r",ax=ax[0])
    plt.suptitle('Countplot')
    sns.countplot(data=df,hue="TARGET",x=feature,linewidth=3,palette="Oranges_r",ax=ax[1])
    plt.suptitle('Countplot')
```

#### Continous Variable

```python
def plot_continuous_univariate(feature):
    fig, ax = plt.subplots(1,2,figsize=(20,5))
    plt.suptitle('DISTRIBUTION PLOTS')
    sns.distplot(df[feature], bins = 50,ax=ax[0])
    plt.suptitle('BOX PLOTS')
    sns.boxplot(df[feature], ax=ax[1]);
```

#### Bivaraiate analysis

```python
def plot_categorical_bivariate(feature):
    fig=plt.figure(figsize=[4,4])
    sns.countplot(data=df,hue="TARGET",x=feature,linewidth=3,palette="Oranges_r")
    plt.xlabel(feature,fontdict={"fontsize":10,'fontweight':5, 'color':"Black"})
    plt.ylabel("COUNT",fontdict={"fontsize":10,'fontweight':5, 'color':"Black"})
    plt.show()
```
